#### **<span style="color:blue">[2-1] air_21, air_22 의 '측정일시'를 활용하여 'time' 변수 생성</span>**

* air_21, air_22 의 '측정일시'를 활용하여 'time'변수 생성
* time 변수를 to_datetime으로 데이터 타입 변경
* 참고: 미세먼지 데이터는 1시-24시, 날씨 데이터는 0시-23시로 구성되어 있습니다. [2-5]에서 미세먼지와 날씨 데이터를 time 기준으로 합치려면 날씨 기준이 동일해야 합니다. 미세먼지 데이터에서 time 변수 생성 시 이를 미리 고려(예:측정일시 값 -1)하세요.

```python
# 아래에 필요한 코드를 작성하고 결과를 확인합니다.

def add_time_column(df):
    df['측정일시'] = df['측정일시'].astype(str)
    df['hour'] = df['측정일시'].str[-2:].astype(int) - 1  # 1시간 보정
    df['date'] = df['측정일시'].str.slice(0, 8)
    df['time'] = pd.to_datetime(df['date'], format='%Y%m%d') + pd.to_timedelta(df['hour'], unit='h')
    df.drop(columns=['hour', 'date'], inplace=True)
    return df

air_21 = add_time_column(air_21)
air_22 = add_time_column(air_22)
```

```python
# 확인 1: time 컬럼과 원본 '측정일시' 일부 확인
print(air_21[['측정일시', 'time']].head())
print(air_22[['측정일시', 'time']].head())

# 확인 2: time 컬럼에서 시간대가 0~23인지 확인
print(air_21['time'].dt.hour.unique())
print(air_22['time'].dt.hour.unique())
```
